```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'age': [25, 45, 56, 32, 47],
    'income': [50000, 64000, 80000, 52000, 200000]
}
df = pd.DataFrame(data)

# Before scaling
print("Before Scaling:")
print(df)

# Applying Min-Max scaling
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

# Create a new dataframe with scaled values
df_scaled = pd.DataFrame(scaled_data, columns=['age', 'income'])

# Visualize before and after scaling
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Before scaling
ax1.scatter(df['age'], df['income'], color='blue')
ax1.set_title('Before Scaling')
ax1.set_xlabel('Age')
ax1.set_ylabel('Income')

# After scaling
ax2.scatter(df_scaled['age'], df_scaled['income'], color='green')
ax2.set_title('After Min-Max Scaling')
ax2.set_xlabel('Age')
ax2.set_ylabel('Income (Scaled)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Sample dataset with outliers
data = {
    'age': [25, 45, 56, 32, 47],
    'income': [50000, 64000, 80000, 52000, 200000]  # Notice the outlier in income
}
df = pd.DataFrame(data)

# Before Standardization
print("Before Standardization:")
print(df)

# Applying Z-Score Standardization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Create a new dataframe with scaled values
df_scaled = pd.DataFrame(scaled_data, columns=['age', 'income'])

# Visualize before and after standardization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Before scaling
ax1.scatter(df['age'], df['income'], color='blue')
ax1.set_title('Before Standardization')
ax1.set_xlabel('Age')
ax1.set_ylabel('Income')

# After standardization
ax2.scatter(df_scaled['age'], df_scaled['income'], color='green')
ax2.set_title('After Z-Score Standardization')
ax2.set_xlabel('Age')
ax2.set_ylabel('Income (Standardized)')

plt.tight_layout()
plt.show()
```
