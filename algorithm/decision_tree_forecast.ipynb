```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('weather_forecast.csv')
df
```

```python
df['Outlook'] = df['Outlook'].replace({'Sunny':1, 'Rain':2, 'Overcast':3})
df['Temperature'] = df['Temperature'].replace({'Hot':1, 'Cool':2, 'Mild':3})
df['Humidity'] = df['Humidity'].replace({'High':1, 'Normal':0})
df['Windy'] = df['Windy'].replace({'Strong':1, 'Weak':0})
df['Play'] = df['Play'].replace({'No':0, 'Yes':1})
```

```python
df
```

```python
X = df.drop('Play', axis=1)
y = df['Play']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
c4_5 = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
c4_5.fit(X, y)
```

```python
plt.figure(figsize=(12,14))
tree.plot_tree(c4_5, feature_names= X.columns.tolist(), class_names=['Bad', 'Good'], filled=True, rounded=True)
plt.savefig('./c4_5_tree_forecast.png', bbox_inches="tight")
```

```python
cart = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
cart.fit(X, y)
y_pred_cart = cart.predict(X)
```

```python
plt.figure(figsize=(12,14))
tree.plot_tree(cart, feature_names= X.columns.tolist(), class_names=['Bad', 'Good'], filled=True, rounded=True)
plt.savefig('./cart_tree_forecast.png', bbox_inches="tight")
```
